## 设置坐标刻度

plt.xticks(刻度, 刻度标签)   
plt.yticks(刻度, 刻度标签)

这里的刻度和刻度标签都是序列（包括元组、列表），因而可以使用:
1.numpy中的np.arange(起始值,终止值,步长) 进行传参
2.Python的内置range(起始值,终止值,步长)  进行传参

刻度指的是分隔点的距离，刻度标签指的是分隔点的名称

```python
# 更改y轴刻度，不然这个图片会让人感觉温差很大（温度区间改成-10到40摄氏度）
# 更改x轴刻度，使得x轴的刻度标签为时间

# 先导包
import random
import matplotlib.pyplot as plt

# 准备数据
x = range(60)
y = [random.uniform(15, 18) for i in x]

# 创建画布（窗口）
plt.figure(figsize=(20, 8), dpi=120)

# 绘制图像
plt.plot(x, y)

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 更改y轴的刻度
plt.yticks(range(-10, 41, 5))

# 制作x轴的刻度标签
time = []
for minute in range(60):
    time.append(f"11点{minute}分")

# 更改x轴的刻度，并添加刻度标签
plt.xticks(x[::5], time[::5])

# 显示图像
plt.show()
```

### 补充：format方法

formatted_string = "Hello, {}".format(name)
{}: 占位符，表示将被 format 方法中的参数替换。

```python
# 基本替换
name = "Alice"
greeting = "Hello, {}".format(name)
print(greeting)  # 输出: Hello, Alice
```

```python
# 多个参数替换
name = 'Tom'
age = 18
info = "{} is {} years old.".format(name, age)
print(info)  # 输出: Tom is 18 years old.
```

```python
# 指定位置
name = 'Tom'
age = 18
info = "{1} is {0} years old.".format(age, name)
print(info)  # 输出: Tom is 18 years old.
```

#### 对多个曲线进行随机颜色分配

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 生成 x 数据
x = np.linspace(0, 20, 10)  # x 从 0 到 20，共 10 个点
num_curves = 5  # 曲线数量

# 2. 随机生成颜色
# 使用 np.random.rand 生成一个形状为 (num_curves, 3) 的数组
# np.random.rand 是 NumPy 中的一个函数，用于生成一个给定形状的数组，
# 数组中的元素是从均匀分布中随机抽取的，范围在 [0, 1) 之间。
# num_curves：表示要生成的曲线数量。
# 3：表示每个颜色由三个值组成，分别对应 RGB 颜色模型中的红色、绿色和蓝色。
colors = np.random.rand(num_curves, 3)  # 创建 num_curves 行，每行 3 列的随机颜色

# 3. 绘制多条曲线
for i in range(num_curves):
    y = x + i  # 生成不同的 y 数据
    plt.plot(x, y, color=colors[i])  # 使用随机颜色绘制曲线

# 4. 添加标题和显示图形
plt.title('随机颜色曲线')
plt.xlabel('X 轴')
plt.ylabel('Y 轴')
plt.grid(True)  # 显示网格
plt.show()  # 显示图形
```

### 原本的城市记为A，添加新的城市B，并对plot方法的内置参数进行演示

## 同一窗口内创建多个坐标系（子图）-plt.subplot(面向对象)

```python
# 先导包
import random
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 准备数据
x = range(60)
y_A = [random.uniform(15, 18) for i in x]
y_B = [random.uniform(10, 14) for i in x]

# 创建画布（窗口）
plt.figure(figsize=(20, 8), dpi=120)

# 创建子图1
plt.subplot(1, 2, 1)

# 绘制图像，演示plot方法内置参数
plt.plot(x, y_A, color='red', linestyle='-', label='A市温度变化图线', marker="o")

# 更改y轴的刻度
plt.yticks(range(-10, 41, 5))

# 制作x轴的刻度标签
time = ["11点{}分".format(minute) for minute in x]

# 更改x轴的刻度，并添加刻度标签
plt.xticks(x[::10], time[::10])

# 添加x、y轴的描述信息
plt.xlabel("时间")
plt.ylabel("温度")

# 添加标题
plt.title("A市中午十一点到十二点的温度变化信息")

# 创建子图2
plt.subplot(1, 2, 2)

# 绘制图像，演示plot方法内置参数
plt.plot(x, y_B, color="blue",linestyle="--", label="B市温度变化图线", marker="^")

# 更改y轴的刻度
plt.yticks(range(-10, 41, 5))

# 制作x轴的刻度标签
time = ["11点{}分".format(minute) for minute in x]

# 更改x轴的刻度，并添加刻度标签
plt.xticks(x[::10], time[::10])

# 添加x、y轴的描述信息
plt.xlabel("时间")
plt.ylabel("温度")

# 添加标题
plt.title("B市中午十一点到十二点的温度变化信息")

# 显示图例
plt.legend()

# 添加网格线
plt.grid(True, alpha=0.6)

# 显示图像
plt.show()
```

### 矩阵式布局和网格式布局

```python
# 网格式布局（支持单元格的合并）
# 先导包
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# 创建窗口
plt.figure("Grid Layout", facecolor='lightgray')

# 调用GridSpec方法
gs = gridspec.GridSpec(2,2)  # 变量gs接收，下面对gs进行划分

# 调用subplot方法进行划分

# 第一行两个进行拼接
plt.subplot(gs[0, :])

# 输入内容
plt.plot([1, 2],[1, 2])

# 第二行第一列
plt.subplot(gs[1, 0])

# 输入内容
plt.plot([1, 2],[2, 4])

# 第二行第二列
plt.subplot(gs[1, 1])

# 输入内容
plt.plot([1, 2],[3, 6])

# 绘图
plt.show()
```

### NumPy.linspace方法（π在numpy中用np.pi得到）

```python
# 先导包
import numpy as np

# 使用linspace语法创建数组
a = np.linspace(1, 5, num = 3)  # 从1到5分成平均的三个数
print(a)

b = np.linspace(1, 5, num = 3, dtype = int)  # dtype改成整数
print(b)

c = np.linspace(0, 2 * np.pi, num = 5)  # np.pi指的就是π，直接用π会报错
print(c)

# 易知分割后的间隔为：(stop值-start值) / (num-1)
# 人为计算间隔难免会有错误，故而使用retstep代码进行更好
d, step = np.linspace(1, 5, num = 3, retstep = True, dtype = int)  # dtype改成整数
print(d)
print(step)

# 注意上面应当用两个变量接收，因为retstep有两个返回值，如果只有一个会报错
try:
    d = np.linspace(1, 5, num = 3, retstep = True, dtype = int)
    print(d)
except Exception as e:
    print(f"报错类型为：{e}")
# 这里居然没报错，目前不知原因
```

### 添加水平线与垂直线

```python
import numpy as np
import matplotlib.pyplot as plt

# 准备数据
x = np.array([1, 2, 3, 4, 5])
y = np.array([10.4, 11.2, 12.1, 10.3, 10.2])

# 绘制初始图
plt.plot(x, y, marker='o')  # 添加 marker 使数据点更明显

# 绘制垂直线 vertical
"""语法：plt.vlines(yvalue, ymin, ymax, ...)  其中yvalue指的是每一个y对应的值（即直线x=c中 c的值，ymin表示起始，ymax表示终止）"""
plt.vlines(3, 10.2, 12.1)
# 绘制水平线 horizontal
"""语法：plt.hlines(xvalue, xmin, xmax, ...)  其中xvalue指的是每一个x对应的值（即直线y=c中 c的值，xmin表示起始，xmax表示终止）"""
plt.hlines(11.5, 1, 5)

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 准备数据
x = np.array([1, 2, 3, 4, 5])
y = np.array([10.4, 11.2, 12.1, 10.3, 10.2])

# 绘制初始图
plt.plot(x, y, marker='o')  # 添加 marker 使数据点更明显

# 绘制多个水平线 horizontal（多个垂直线vertical同）
"""语法：plt.hlines(xvalue, xmin, xmax, ...)  其中xvalue可以使用列表表示多个"""
plt.hlines([11.5, 11], 1, 5)

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 准备数据（为了方便演示改了y的值）
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50])

# 绘制初始图
plt.plot(x, y, marker='o')  # 添加 marker 使数据点更明显

# 绘制分段水平线 horizontal（分段垂直线vertical同）
"""语法：plt.hlines(xvalue, xmin, xmax, ...)  其中xmin,xmax也可以使用列表划分多个"""
plt.hlines([10, 20, 30], [1, 2, 3], [2, 3, 5])  # 1-2对应10, 2-3对应20， 3-5对应30

# 显示图表
plt.show()
```

### 管理边框属性

```python
# 先导包
import matplotlib.pyplot as plt

# 准备数据
x = [-1, 0, 1]
y = [-2, 0, 2]

# 绘制图像
plt.plot(x, y, label="y=2x")

# 使y=2x通过原点
ax = plt.gca()  # 获取当前坐标轴
ax.spines['left'].set_position('zero')  # 左轴位置
ax.spines['bottom'].set_position('zero')  # 底轴位置
ax.spines['right'].set_color('none')  # 隐藏右轴
ax.spines['top'].set_color('none')  # 隐藏上轴

# 显示图例
plt.legend()

# 显示图像
plt.show()
```
