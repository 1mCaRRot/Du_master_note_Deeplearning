# 1.折线图（plot）

## 1.1绘制正弦函数图像

```python
# 先导包
import matplotlib.pyplot as plt
import numpy as np

# 生成 x 数据
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

# 绘制曲线
plt.plot(x, y)

# 设置坐标刻度
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r'-$\pi$', r'-$\frac{\pi}{2}$', r'0', r'$\frac{\pi}{2}$', r'$\pi$'])  # 使用 LaTeX 表示分数
plt.yticks((-1, -0.5, 0.5, 1))  # 因为这里的 坐标刻度 传入序列即可，故演示使用元组进行传入

# 添加标题和标签
plt.title('正弦曲线')

# 设置坐标轴位置
ax = plt.gca()  # 获取当前坐标轴
ax.spines['left'].set_position('zero')  # 左轴位置
ax.spines['bottom'].set_position('zero')  # 底轴位置
ax.spines['right'].set_color('none')  # 隐藏右轴
ax.spines['top'].set_color('none')  # 隐藏上轴

# 添加网格
plt.grid(True)

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 显示图形
plt.show()
```

### 1.1.2 标记特殊点，并添加注释（使用annotate语法）

```python
# arrowprops 总结

# 1. 定义
# arrowprops 是 plt.annotate 函数中的一个参数，用于设置箭头的属性。

# 2. 数据结构
# arrowprops 是一个字典
arrowprops_dict = dict(facecolor='black', arrowstyle='->', lw=2)  # 使用 dict() 函数
arrowprops_alternative = {'facecolor': 'black', 'arrowstyle': '->', 'lw': 2}  # 替代写法

# 3. 常见属性
# - facecolor: 箭头的填充颜色
# - edgecolor: 箭头边缘的颜色
# - arrowstyle: 箭头的样式
# - lw: 箭头的线宽

# 4. 示例
import matplotlib.pyplot as plt
import numpy as np

# 生成数据
x = np.linspace(-np.pi, np.pi, 1000)
y = np.sin(x)

# 绘制曲线
plt.plot(x, y)

# 使用 arrowprops 传递字典
plt.annotate('Special Point', 
             xy=(np.pi/2, 1), 
             xytext=(np.pi/2 + 0.5, 1.5), 
             arrowprops=arrowprops_dict)

# 如果要标记特殊点，可以使用scatter语法，如下：
plt.scatter([np.pi/2], [1], color='red',marker='^')  # 标记特殊点

# 显示图形
plt.show()
```

### 1.1.3 使用上述方法对1.1的正弦函数进行标记并注释

```python
# 先导包
import matplotlib.pyplot as plt
import numpy as np

# 生成 x 数据
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

# 绘制曲线
plt.plot(x, y)

# 设置坐标刻度
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r'-$\pi$', r'-$\frac{\pi}{2}$', r'0', r'$\frac{\pi}{2}$', r'$\pi$'])
plt.yticks((-1, -0.5, 0.5, 1))

# 添加标题和标签
plt.title('正弦曲线')

# 设置坐标轴位置
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 添加网格
plt.grid(True)

# 标注特殊点 (π/2, 1)
plt.scatter([np.pi/2], [1], color='red')  # 标记特殊点

# 添加特殊点注释
plt.annotate(r'$\left(\frac{\pi}{2}, 1\right)$', # 备注中显示的文本内容
             xy=(np.pi/2, 1),    # 备注目标点的坐标
             xytext=(np.pi/2 + 0.2, 1.2),  # 特殊点注释文本的位置
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# 显示图形
plt.show()
```

## 3.2 绘制各电影在首日和首周的票房统计柱状图

```python
# 先导包
import matplotlib.pyplot as plt
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 准备数据
movie_names = ["A", "B", "C", "D"]
tickets_firstday = [14589, 21444, 2354, 1245]
tickets_firstweek = [73853, 57767, 22354, 8234]

# 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 分配x轴刻度（要把两个图形的x刻度分开）
movie_num_list = np.arange(len(movie_names))
x_firstday = [i - 0.1 for i in movie_num_list]
x_firstweek = [i + 0.1 for i in movie_num_list]

# 绘制柱状图
plt.bar(x_firstday, tickets_firstday, label="首日票房", width=0.2, color="blue")
plt.bar(x_firstweek, tickets_firstweek, label="首周票房", width=0.2, color="red")

# 添加x轴刻度标签
plt.xticks(movie_num_list, ["A", "B", "C", "D"])

# 添加标题
plt.title("某影院各电影票房统计")

# 添加图例
plt.legend()

# 显示图像
plt.show()
```

# 4.直方图（histogram）

#### 组数（bins） = 极差 / 组距 = （max - min） / distance

## 4.1 绘制电影时长分布直方图

```python
# 先导包
import random
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 准备数据，这里用random.randint方法得到500个在90-140之间的整数
movie_time = [random.randint(90, 140) for i in range(500)]

# 创建画布（窗口）
plt.figure(dpi=120, figsize=(20, 8))

# 计算极差（range）,并确立组距(distance)，以明确我们需要填入的 hist参数中的 组数（bins）
range_value = max(movie_time) - min(movie_time)
distance_value = 2
bins_value = int(range_value / distance_value)   # 这里应当确保bins是整数，如果是浮点数则报错

# 绘制直方图（hist）
plt.hist(movie_time, bins_value)

# 调整刻度及刻度标签
plt.xticks(range(min(movie_time), max(movie_time) + 1, distance_value))

# 添加坐标轴备注
plt.xlabel("电影时长")
plt.ylabel("个数")
plt.title("电影时长分布直方图")

# 显示网格线
plt.grid(linestyle="--")

# 显示图像
plt.show()
```

# 5.饼状图（pie）

## 5.1 绘制水果销售的统计饼状图

```python
# 先导包
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 创建画布（窗口）
plt.figure(figsize=(20, 8), dpi=120)

# 准备数据
fruit = [15, 30, 45, 10]  # 各部分的大小
fruit_names = ['苹果', '香蕉', '橙子', '火龙果']  # 各部分的标签
fruit_colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']  # 各部分的颜色

# 绘制饼图
plt.figure(figsize=(8, 8))
plt.pie(fruit, labels=fruit_names, colors=fruit_colors, autopct='%1.1f%%', startangle=0)

# 添加图例
plt.legend()

# 添加标题
plt.title("")

# 显示图像
plt.axis('equal')  # 保证饼图是圆形
plt.show()
```
