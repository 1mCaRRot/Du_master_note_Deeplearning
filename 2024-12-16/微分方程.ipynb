```julia
using Plots

# 定义参数
a = 1.5   # 参数 a
b = 0.02  # 参数 b
c = 0.002 # 参数 c
d = 0.6   # 参数 d

# 初始条件
x1 = 250.0 # x1(0) = 250
x2 = 100.0 # x2(0) = 100
dt = 0.01  # 时间步长
t = 0.0    # 初始时间

# 用于存储数据
times = []
x1_values = []
x2_values = []

# 定义时间步进函数
function step()
    global x1, x2, t  # 声明这些变量为全局变量，以便在函数内部修改它们
    dx1 = a * x1 - b * x1 * x2  # x1' = ax1 - bx1x2
    dx2 = c * x1 * x2 - d * x2  # x2' = cx1x2 - dx2
    x1 = x1 + dx1 * dt
    x2 = x2 + dx2 * dt
    t = t + dt
end

# 模拟过程
while t < 15.0
    step()
    push!(times, t)
    push!(x1_values, x1)
    push!(x2_values, x2)
end

# 绘图
plot(times, x1_values, label="x1", xlabel="Time", ylabel="Values", title="Dynamics of x1 and x2", legend=:topleft)
plot!(times, x2_values, label="x2")
```
